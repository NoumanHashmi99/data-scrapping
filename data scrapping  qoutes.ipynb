```python
t=soup.find_all('div',class_='tags')
```

```python
tt = [i.text.strip() for i in t]
```

```python
tt
```

```python
bad_char = ['Tags:\n',"'",'\n',"'"]
```

```python
def clean(value):
    for i in bad_char:
        value = value.replace(i,' ')
    return value
```

```python
clean(tt[0])
```

```python
lst = []
for i in tt:
    lst.append(clean(i))
    

    
    
    
    
```

```python
lst
```

```python
df_1['Tags'] = pd.DataFrame(lst)
```

```python
df_1['Tags']=df_1['Tags'].str.replace(' ',',').str.strip(',')
```
